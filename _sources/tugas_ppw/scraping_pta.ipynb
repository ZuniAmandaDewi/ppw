## Import Library

**BeautifulSoup** : digunakan untuk mengurai konten HTML



```python
import requests as req
from bs4 import BeautifulSoup as bs
import csv
```

## Scraping

untuk melakukan proses scraping disini membuat sebyah fungsi bernama scrap_pta, berikut langkah-langkah didalamnya :
*   data=[] : Deklarasi variabel data untuk menampung semua hasil scraping
*   Buat perulangan untuk menentukan halaman ke berapa yang akan dilakukan scraping
*   Variabel perid : digunakan untuk mengambil data per judulnya
*   Dari elemen 'li' yang ada di dalam variabel perid juga terdapat elemen 'a' dengan atribut 'href' untuk mengambil link abstrak tersebut.
*   Link yang sudah diambil kemudian dijalankan untuk mengambil nama penulis pada variabel penulis_elem. Teks yang terkandung dalam variabel penulis_elem dipotong dari karakter ke-11 dan disimpan dalam variabel penulis.
*   Pengambilan data dosen pembimbing berada pada elemen div, teks yang terkandung dalam elemen span. data yang dihasilkan pada variabel dospem_elem akan dilakukan penghapusan kata 'Dosen Pembimbing I' dan 'Dosen Pembimbing II'
*   pengambilan abstrak : Teks yang terkandung dalam elemen p yang ditemukan disimpan dalam variabel abst. Teks yang terkandung dalam variabel abst diubah sehingga karakter \r dan \n diganti dengan karakter kosong dan disimpan kembali dalam variabel
*   pengambilan judul : Teks yang terkandung dalam elemen a yang ditemukan disimpan dalam variabel title. Teks yang terkandung dalam variabel title diubah sehingga karakter \r dan \n diganti dengan karakter kosong dan disimpan kembali dalam variabel
*   Semua data abstrak, judul, penulis, dosen pembimbing 1, dosen pembimbing 2 disimpan dalam variabel data kemudian dirubah kedalam bentuk dataframe dan disimpan ke drive.












```python
def scrap_pta(hal):
  data=[]
  for page in range(1,hal):
        url = f"https://pta.trunojoyo.ac.id/c_search/byprod/10/{page}"
        request = req.get(url).text
        soup = bs(request, 'html.parser')
        perid = soup.find_all('li', {'data-id':'id-1'})
        for x in perid:
          link = x.find('a','gray button')['href']
          request2 = req.get(link).text
          soup2 = bs(request2, 'html.parser')
          penulis_elem = soup2.find('div', {'style': 'padding:2px 2px 2px 2px;'}).find('span').text
          penulis = penulis_elem[10:]
          dospem_elem = soup2.find('div', {'style': 'float:left; width:540px;'}).findAll('div', {'style': 'padding:2px 2px 2px 2px;'})
          for dospem in dospem_elem:
            dospem_text = dospem.find('span').text
            if 'Dosen Pembimbing I :' in dospem_text:
                dospem_i = dospem_text.replace('Dosen Pembimbing I :', '').strip()
            elif 'Dosen Pembimbing II :' in dospem_text:
                dospem_ii = dospem_text.replace('Dosen Pembimbing II :', '').strip()
          abst= soup2.find('p',{'align':'justify'}).text.replace('\r','').replace('\n','')
          headline = x.find('a', 'title').text.replace('\r','').replace('\n','')
          data.append([headline,penulis,dospem_i,dospem_ii,abst])
          kepala = ['Judul','Penulis','Dosen Pembimbing 1', 'Dosen Pembimbing 2','Abstrak']
          writer = csv.writer(open('/content/drive/MyDrive/ppw/tugas/tugas_ppw/pta.csv', 'w', newline=''))
          writer.writerow(kepala)
          for datas in data :
            writer.writerow(datas)
```

```python
scrap_pta(172)
```
